```python
plt.figure(dpi=200)
plt.hist(np.random.normal(140, 20, 1000), ec='salmon', color='cyan', bins=30)
ax = plt.gca()
for sp, p in zip(['top', 'right'], ['bottom', 'left']):
    ax.spines[sp].set_visible(False)
    ax.spines[p].set_color('salmon')
plt.rcParams['xtick.color'] = 'salmon'
plt.rcParams['ytick.color'] = 'salmon'
plt.rcParams['axes.titlecolor'] = 'red'
plt.rcParams['axes.titlesize'] = 25
plt.rcParams['axes.labelcolor'] = 'salmon'
plt.rcParams['axes.labelsize'] = 20
plt.xticks(range(80, 220, 10), color='salmon', rotation=90)
plt.yticks(range(0, 110, 10), color='salmon')
plt.grid(color='salmon', alpha=0.7, ls=':')
plt.xlabel('Height')
plt.ylabel("Frequency")
plt.title("A Normal Distributed Data of Height", pad=50)
plt.show()
```
